```python
data_train = pd.read_csv('data_train.csv')
data_test = pd.read_csv('data_test.csv')
```

```python
data_train
```

```python
column_names = data_train.select_dtypes(exclude = 'number').columns
```

```python
for col in column_names:
    data_train[col][data_train[col] == 'yes'] = 1
    data_train[col][data_train[col] == 'no'] = 0
```

```python
data_train
```

```python
for col in column_names:
    data_test[col][data_test[col] == 'yes'] = 1
    data_test[col][data_test[col] == 'no'] = 0
```

```python
data_train
```

```python
data_train['label'].count()
```

```python
data_train.info()
```

```python
prop = round(data_train['account'][data_train['label'] == 1].count()/data_train['label'].count(), 3)
```

```python
onehot_train = data_train
```

```python
onehot_train = onehot_train.drop(columns = ['label'])
```

```python
onehot_train
```

```python
onehot_test = data_test.drop(columns = ['label'])
```

```python
onehot_test
```

```python
y_train = data_train['label']
```

```python
onehot_train.info()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(max_depth = 10, random_state = 0, n_estimators = 30, class_weight = {0: prop, 1: 1-prop})
rf_model.fit(onehot_train, y_train)
```

```python
rf_model.feature_importances_
```

```python
from sklearn.metrics import recall_score
```

```python
y_pred = rf_model.predict(onehot_test)
```
